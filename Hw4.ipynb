<h2>Problem 0</h2>

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

<h2>Problem 1</h2>

1a. <br>
Make a scatterplot of the first 1000 data.

```python
plt.scatter(t[0:999], bkgd[0:999])
plt.show()
```

1b.

```python
numpoints = 2000
startpoint = 500
plt.scatter(t[0:numpoints], bkgd[startpoint - 1:startpoint + numpoints - 1])
plt.show()
```

1c.

```python
numpoints = int(len(bkgd) / 100)

plt.scatter(t[0:numpoints], bkgd[::100])
plt.show()
```

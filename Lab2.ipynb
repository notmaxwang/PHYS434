```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

#Problem 1

plt.rcParams["figure.figsize"] = (15,10)
fig, ax = plt.subplots(3, 2)
plt.xlim([0,10])

x = np.arange(0,10)
y = stats.poisson.pmf(x,mu = 6)
ax[0,0].plot(x,y)
ax[0,0].set_xlabel("Number of Rays");
ax[0,0].set_ylabel("Probability");
ax[0,0].set_title("Day1");

y2 = np.convolve(y, y);
x2 = np.arange(0, 19)
ax[0,1].plot(x2,y2);
ax[0,1].set_title("Day2");

y3 = np.convolve(y2, y);
x3 = np.arange(0,28)
ax[1,0].plot(x3, y3);
ax[1,0].set_title("Day3");

y4 = np.convolve(y3, y);
x4 = np.arange(0,37)
ax[1,1].plot(x4, y4);
ax[1,1].set_title("Day4");

y5 = np.convolve(y4, y);
x5 = np.arange(0, 46)
ax[2,0].plot(x5, y5);
ax[2,0].set_title("Day5");
plt.show()
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

plt.rcParams["figure.figsize"] = (15,10)
fig, ax = plt.subplots(3, 2)
plt.xlim([0,10])

x = np.arange(0,10)
y = stats.rayleigh.pdf(x,loc = 6)
ax[0,0].plot(x,y)
ax[0,0].set_xlabel("Number of Rays");
ax[0,0].set_ylabel("Probability");
ax[0,0].set_title("Day1");

y2 = np.convolve(y, y);
x2 = np.arange(0, 19)
ax[0,1].plot(x2,y2);
ax[0,1].set_title("Day2");

y3 = np.convolve(y2, y);
x3 = np.arange(0,28)
ax[1,0].plot(x3, y3);
ax[1,0].set_title("Day3");

y4 = np.convolve(y3, y);
x4 = np.arange(0,37)
ax[1,1].plot(x4, y4);
ax[1,1].set_title("Day4");

y5 = np.convolve(y4, y);
x5 = np.arange(0, 46)
ax[2,0].plot(x5, y5);
ax[2,0].set_title("Day5");

plt.show()
```

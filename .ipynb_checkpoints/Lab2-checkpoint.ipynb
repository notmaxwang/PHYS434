```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

```

Part a.
The distribution is set up using the stats.poisson.pmf package, and the noise of the background of these gamma-ray sources is 
set to have values from 0 to 10, and the average is set to be 6. The probability of a sum of the distribution can be calculated
with convolution, and it is set up using the np.convolve package to inteagrate the probability mass function of the past day with 
the current day, and continue to get the results for more days. 


```python
plt.rcParams["figure.figsize"] = (15,20)
fig, ax = plt.subplots(3, 2)

x = np.arange(0,10)
y = stats.poisson.pmf(x,mu = 6)
ax[0,0].plot(x,y)
ax[0,0].set_xlabel("Cosmic Ray Noise");
ax[0,0].set_ylabel("Probability");
ax[0,0].set_title("Day1");

y2 = np.convolve(y, y);
x2 = np.arange(0, 19)
ax[0,1].plot(x2,y2);
ax[0,1].set_xlabel("Cosmic Ray Noise");
ax[0,1].set_ylabel("Probability");
ax[0,1].set_title("Day2");

y3 = np.convolve(y2, y);
x3 = np.arange(0,28)
ax[1,0].plot(x3, y3);
ax[1,0].set_xlabel("Cosmic Ray Noise");
ax[1,0].set_ylabel("Probability");
ax[1,0].set_title("Day3");

y4 = np.convolve(y3, y);
x4 = np.arange(0,37)
ax[1,1].plot(x4, y4);
ax[1,1].set_xlabel("Cosmic Ray Noise");
ax[1,1].set_ylabel("Probability");
ax[1,1].set_title("Day4");

y5 = np.convolve(y4, y);
x5 = np.arange(0, 46)
ax[2,0].plot(x5, y5);
ax[2,0].set_xlabel("Cosmic Ray Noise");
ax[2,0].set_ylabel("Probability");
ax[2,0].set_title("Day5");

y6 = np.convolve(y5, y);
x6 = np.arange(0, 55)
ax[2,1].plot(x6, y6);
ax[2,1].set_xlabel("Cosmic Ray Noise");
ax[2,1].set_ylabel("Probability");
ax[2,1].set_title("Day6");

plt.show()
```

From the results, it shows that the more days there is, the higher the mean is, and the smaller the smaller the standard error is.


Part B.
This is a comparison of the sum of over five days of the distribution, with
a generated Poisson distribution. The Red line is the generated Poisson distribution
and the blue line is the day five sum of the distribution.

```python
plt.rcParams["figure.figsize"] = (15,10)
fig, ax = plt.subplots(1, 1)
ax.set_title("Day 5 Sum VS. Generated Poisson Distribution")
ax.set_xlabel("Cosmic Ray Noise")
ax.set_ylabel("Probability")
ax.plot(x5, y5);
ax.plot(x5, stats.poisson.pmf(x5, 6*5), color = "r")

plt.show()
```

Part C.
This is the probability distribution as each is averaged. The sum is taken from part
A and it is divided by the day passed to get the average of each day. 

```python
plt.rcParams["figure.figsize"] = (15,10)
fig, ax = plt.subplots(1, 1)
ax.plot(x, y);
ax.plot(x2 / 2, y2);
ax.plot(x3 / 3, y3);
ax.plot(x4 / 4, y4);
ax.plot(x5 / 5, y5);
ax.plot(x6 / 6, y6);
ax.set_title("Poisson Average Distribution")
ax.set_xlabel("Cosmic Ray Noise")
ax.set_ylabel("Probability")

plt.show()
```

The average mean of the distribution stays the same, and the main change is the standard deviation of 
all of the distribution is smaller as the day goes on. 

Part D.
The number of days will be 6 days, and Y is 8. The sigma is calcualted by first accumulating
the noise that is bigger than 8 * 6 which is 48, and to find the sigma the function 
stats.norm.ppf is used, which is 1 - the accumulated probability. The final sigma is 
around 3.51.

```python

prob = 0
for i, val in enumerate(y6):
    if (i > 8*6):
        prob += val
        
stats.norm.ppf(1-prob)
```

Part A.
The distribution is set up using the function stats.rayleigh.pdf, and this set up the
sum of the distribution over multiple days.

```python
plt.rcParams["figure.figsize"] = (10,10)
fig, ax = plt.subplots(5, 1)

x = np.arange(0,10)
y = stats.rayleigh.pdf(x,loc = 6)
ax[0].plot(x,y)
ax[0].set_xlabel("Number of Rays");
ax[0].set_ylabel("Probability");
ax[0].set_title("Day1");

y2 = np.convolve(y, y);
x2 = np.arange(0, 19)
ax[1].plot(x2,y2);
ax[1].set_title("Day2");

y3 = np.convolve(y2, y);
x3 = np.arange(0,28)
ax[2].plot(x3, y3);
ax[2].set_title("Day3");

y4 = np.convolve(y3, y);
x4 = np.arange(0,37)
ax[3].plot(x4, y4);
ax[3].set_title("Day4");

y5 = np.convolve(y4, y);
x5 = np.arange(0, 46)
ax[4].plot(x5, y5);
ax[4].set_title("Day5");

plt.show()
```

This sets up the average 

```python
plt.rcParams["figure.figsize"] = (15,10)
fig, ax = plt.subplots(1, 1)
ax.plot(x, y);
ax.plot(x2 / 2, y2);
ax.plot(x3 / 3, y3);
ax.plot(x4 / 4, y4);
ax.plot(x5 / 5, y5);

plt.show()
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy import stats
import h5py
import pandas as pd
plt.rcParams["figure.figsize"] = (10,8)
```

<h3>Problem 1</h3>

```python

bkgd = stats.norm.rvs(loc=0,scale=2, size = 1000000)
signal = stats.uniform.rvs(loc=0,scale=20, size = 1000000)

data = bkgd + signal

signaledges = np.linspace(0,20,40)
dataedges = np.linspace(-7,27,68)

Psd, temp, temp2= np.histogram2d(data,signal, bins=[dataedges,signaledges], density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.title("Histogram of Observed Signal vs True Signal")
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
plt.show()
```

The background distribution is centered at zero, with the sigma value of 2, and the <br>
signal has a sigma value of 20. 

<h3>1b.</h3>

The injected signal that I've chosen is set to be 15. 

```python

plt.step(datacenters,Psd[:,30])
plt.yscale('log')
plt.xlim([0,30])
plt.title('True Signal = 15')
plt.xlabel('Observed Signal')
plt.ylabel('Probability')
plt.show()
```

The above distribution looks somewhat Gaussian, with the mean being the<br>
true injected signal value at 15.  This shows the possible distribution <br>
of the observed signals and from the plot, and it shows the possible signal <br>
from -5 to 19.

<h3>1c.</h3>

```python

plt.step(signalcenters,Psd[30,:])
plt.yscale('log')
plt.xlim([0,20])
plt.xlabel('True Signal')
plt.ylabel('Probability')
plt.title('Observed Signal = 8')
plt.show()
```

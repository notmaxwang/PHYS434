```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

import pickle
```

In lab 5, we are expolring the data for the standard model Higgs boson in proton to proton <br>
collisions at the center of mass energy of 13 TeV. This data is simulated from the Large Hadron <br>
Collider, which is 27 kilometer in circumference, and it accelerate protons against each other. <br>
The collider has seven detectors, and when these protons collide, the data can provide an insight <br>
into what the proton is made out of, and it creates quantum excitation in the higgs field, which <br>
helps explore the Higgs boson.

This lab is focused on one section of the LHC, which is the ATLAS(A toroidal LHC Apparatus)<br>
which has a main objective to confirm the existance of the Higgs boson. The major charstistic <br>
of  the Higgs boson is its large mass compared to other subatomic particles and the short <br>
halflife. Since the given collision energy is 13TeV, the resulting particles from the Higgs boson<br>
collision will decay into a bottom quark-antiquark pair, two photons, and two bosons. And we will<br>
use the provided dataset, the Higgs data and the QCD (quantum chromodynamic contamination)  <br>
data to determine this. 

```python
# open the file of interest, and use pickle loading
higgs_data = pickle.load(open ("datafile/higgs_100000_pt_250_500.pkl",'rb'))
qcd_data = pickle.load(open ("datafile/qcd_100000_pt_250_500.pkl",'rb'))
```

What are some of these data sets categories? <br>

<li>pt - transverse momentum in the xy plane of the pp collisons. $\sqrt{p_x^2 + p_y^2}$</li>
<li>eta - pseudorapidity, which is the angle of a particle relative to the beam axis.$ - \ln( \tan(\frac{\theta}{2}))$ </li>
<li>phi - azimuthal angle, which is the angle above the x axis. $\cos^{-1}\frac{x}{r} $</li>
<li>mass - invariant mass, or the rest mass of the particle system $ E^2=p^2+m^2 $</li>
<li>ee2 - 2-point ECF ratio, which is the energy correlation function ratio for 2-point correlators</li>
<li>d2 - ratio of 3 to 2 point ECF</li>
<li>angularity - orientation of one feature to another at a referenced angle</li>
<li>t1 - 1-subjettiness, which is a signiture of quark or gluon like object created by the collision</li>
<li>KtDeltaR - delta R of two subjets within the large_R jet</li>

<h4>Histogram of Transverse momentum</h4>

```python
plt.rcParams["figure.figsize"] = (30,20)  
fig, ax = plt.subplots(2, 2)

ax[0, 0].hist(higgs_data.get('pt'), density = True, bins = 50, alpha = 0.6)
ax[0, 0].set_title('Transverse Momentum for Higgs data set')
ax[0, 0].set_xlabel('Transverse Momentum')
ax[0, 0].set_ylabel('Frequency')

ax[0, 1].hist(higgs_data.get('pt'), density = True, bins = 50, alpha = 0.8)
ax[0, 1].set_title('Log of Transverse Momentum for Higgs data set')
ax[0, 1].set_yscale('Log')
ax[0, 1].set_xlabel('Transverse Momentum')
ax[0, 1].set_ylabel('log(Frequency)')

ax[1, 0].hist(qcd_data.get('pt'), density = True, bins = 50, color = 'r', alpha = 0.6)
ax[1, 0].set_title('Transverse Momentum for QCD data set')
ax[1, 0].set_xlabel('Transverse Momentum')
ax[1, 0].set_ylabel('Frequency')

ax[1, 1].hist(qcd_data.get('pt'), density = True, bins = 50, color = 'r', alpha = 0.8)
ax[1, 1].set_title('Log of Transverse Momentum for QCD data set')
ax[1, 1].set_yscale('log')
ax[1, 1].set_xlabel('Transverse Momentum')
ax[1, 1].set_ylabel('log(Frequency)')
plt.show()
```

The transverse momentum for both the Higgs and QCD data seem to be skewed to the left, and <br>
from the log scaled graph comparison, the distribution seem to be very similar from 250 to <br>
410 $\frac{GeV}{c}$. That shows that there is discrimination power from 410 $\frac{GeV}{c}$ to <br>
the higher values. 

<h4>Histogram of Rest Mass</h4>

```python
plt.rcParams["figure.figsize"] = (30,20)  
fig, ax = plt.subplots(2, 2)

ax[0, 0].hist(higgs_data.get('mass'), density = True, bins = 50, alpha = 0.6)
ax[0, 0].set_title('Rest Mass for Higgs data set')
ax[0, 0].set_xlabel('Rest Mass')
ax[0, 0].set_ylabel('Frequency')

ax[0, 1].hist(higgs_data.get('mass'), density = True, bins = 50, alpha = 0.8)
ax[0, 1].set_title('Log of Rest Mass for Higgs data set')
ax[0, 1].set_yscale('Log')
ax[0, 1].set_xlabel('Rest Mass')
ax[0, 1].set_ylabel('log(Frequency)')

ax[1, 0].hist(qcd_data.get('mass'), density = True, bins = 50, color = 'r', alpha = 0.6)
ax[1, 0].set_title('Rest Mass for QCD data set')
ax[1, 0].set_xlabel('Rest Mass')
ax[1, 0].set_ylabel('Frequency')

ax[1, 1].hist(qcd_data.get('mass'), density = True, bins = 50, color = 'r', alpha = 0.8)
ax[1, 1].set_title('Log of Rest Mass for QCD data set')
ax[1, 1].set_yscale('log')
ax[1, 1].set_xlabel('Rest Mass')
ax[1, 1].set_ylabel('log(Frequency)')
plt.show()
```

There is a clear distinction between the rest mass of the QCD data set and <br>
the rest mass of the Higgs boson, which is expected as the Higgs rest mass <br>
are a lot higher than the other subatomic particles. From the log scaled <br>
plot there seem to be a clear spike in the 130 $\frac{GeV}{c}$ mark. 

<h4>Histogram of 2-point ECF</h4>

```python
plt.rcParams["figure.figsize"] = (30,20)  
fig, ax = plt.subplots(2, 2)

ax[0, 0].hist(higgs_data.get('ee2'), density = True, bins = 50, alpha = 0.6)
ax[0, 0].set_title('2-Point ECF Ratio for Higgs data set')
ax[0, 0].set_xlabel('2-Point ECF Ratio')
ax[0, 0].set_ylabel('Frequency')

ax[0, 1].hist(higgs_data.get('ee2'), density = True, bins = 50, alpha = 0.8)
ax[0, 1].set_title('Log of 2-Point ECF Ratio for Higgs data set')
ax[0, 1].set_yscale('Log')
ax[0, 1].set_xlabel('2-Point ECF Ratio')
ax[0, 1].set_ylabel('log(Frequency)')

ax[1, 0].hist(qcd_data.get('ee2'), density = True, bins = 50, color = 'r', alpha = 0.6)
ax[1, 0].set_title('2-Point ECF Ratio for QCD data set')
ax[1, 0].set_xlabel('2-Point ECF Ratio')
ax[1, 0].set_ylabel('Frequency')

ax[1, 1].hist(qcd_data.get('ee2'), density = True, bins = 50, color = 'r', alpha = 0.8)
ax[1, 1].set_title('Log of 2-Point ECF Ratio for QCD data set')
ax[1, 1].set_yscale('log')
ax[1, 1].set_xlabel('2-Point ECF Ratio')
ax[1, 1].set_ylabel('log(Frequency)')
plt.show()
```

Looking at the 2-point ECF distribution, it seems like for Higgs dataset, there is a <br>
peak in energy correlation function being around the 0.15 $\frac{GeV}{c}$. The QCD <br>
dataset has a peak around the 0.05 $\frac{GeV}{c}$, which is significally different. <br>
